```python
import pandas as pd
import cv2
import os

from google.colab.patches import cv2_imshow
```

```python
img = cv2.imread('FER/angry/Training_88391908.jpg')
img_gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

cv2_imshow(img_gray)
print(img_gray.shape)
```

```python
classes = ['neutral', 'happy', 'angry', 'sad']
```

```python
data = []
labels = []

for cls in classes:
    for filename in os.listdir('FER/'+cls):
        print('FER/'+cls+'/'+filename)
        if "jpg" in filename:
            img = cv2.imread('FER/'+cls+'/'+filename)
            img_gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
            img_gray = img_gray.flatten()
            data.append(img_gray)
            labels.append(cls)
```

```python
image_data = pd.DataFrame(data)
image_data.shape
```

```python
labels_data = pd.DataFrame(labels, columns=['class'])
labels_data.head()
```

```python
df = pd.concat((image_data, labels_data), axis=1)
df.to_csv('FER_dataset.csv', index=False)
```
