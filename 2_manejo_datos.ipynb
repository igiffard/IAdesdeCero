# Booleano
¿Qué es un booleano?

Un booleano es un tipo de dato que solo puede tomar dos valores:

True → verdadero

False → falso

Se usan para tomar decisiones dentro de un programa.

```python
x = True
y = False

print(x, y)
```

## Operadores lógicos

Los booleanos pueden combinarse con:
| Operador | Significado                            | Ejemplo          | Resultado |
| -------- | -------------------------------------- | ---------------- | --------- |
| `and`    | Ambas condiciones deben ser verdaderas | `True and False` | False     |
| `or`     | Al menos una condición verdadera       | `True or False`  | True      |
| `not`    | Invierte el valor                      | `not True`       | False     |
|   >     | Mayor que                               |  12 > 13 | False |
|   ==    | Igual que                               |  15 == 12 | True |
|   <     | Menor que                               |  12 < 12 | False |
|   <=     | Menor igual que                        |  12 <= 12 | True |
|   >=     | Mayor igual que                        |  15 >= 12 | True |


```python
edad = 20

es_mayor = edad >= 18
tiene_ine = True

puede_entrar = es_mayor and tiene_ine
print(puede_entrar)   # True
```

# Números enteros y **flotantes**
¿Qué es un dato numérico?

En programación, los números se guardan en variables.
Existen dos tipos muy usados:

| Tipo    | Nombre  | Ejemplo            |
| ------- | ------- | ------------------ |
| Entero  | `int`   | 1, 20, 105, -3     |
| Decimal | `float` | 1.5, 3.1416, -0.25 |


## Crear variables numéricas en Python

```python
entero = 10
decimal = 3.5

print(entero)
print(decimal)
```

## Ver el tipo de dato

```python
print(type(entero))
print(type(decimal))
```

## Operaciones básicas con números

```python
a = 10
b = 3.5

suma = a + b
resta = a - b
multiplicacion = a * b
division = a / b

print("Suma:", suma)
print("Resta:", resta)
print("Multiplicación:", multiplicacion)
print("División:", division)
```

## División entera y residuo

```python
division_entera = a // 3
residuo = a % 3

print("División entera:", division_entera)
print("Residuo:", residuo)
```

### Ejemplo:
* 10 dividido entre 3 es 3 con residuo 1 → división entera = 3, residuo = 1

## Potencias y raíces

```python
potencia = a ** 2
raiz = a ** 0.5

print("Potencia:", potencia)
print("Raíz:", raiz)
```

## Redondeo

```python
numero = 3.1415926

print(round(numero))      # redondeo normal
print(round(numero, 2))   # redondeo a 2 decimales
```

## Librería math
La librería math agrega funciones matemáticas más avanzadas que Python trae por defecto.

```python
import math
```

### Raíces cuadradas y potencias

```python
numero = 25

raiz = math.sqrt(numero)      # raíz cuadrada
potencia = math.pow(numero, 2) # elevar al cuadrado

print("Raíz:", raiz)
print("Potencia:", potencia)
```

### Redondeos y funciones útiles

```python
x = 3.7

print(math.floor(x))   # redondea hacia abajo → 3
print(math.ceil(x))    # redondea hacia arriba → 4
print(math.pi)         # valor de pi
```

### Funciones trigonométricas

```python
angulo = math.radians(90)     # convertir grados a radianes
print(math.sin(angulo))       # seno de 90°
```

# Strings (texto)

¿Qué es un string?

Un string es texto: palabras, frases, letras, números escritos como texto.

```python
texto = "Hola Mundo"
print(texto)
type(texto)
```

### Crear strings
Se pueden usar comillas dobles " " o simples ' '

```python
nombre = "Joan"
ciudad = 'Ensenada'
frase = "La inteligencia artificial es divertida."
```

### Concatenar (unir texto)

```python
nombre = "Joan"
mensaje = "Hola " + nombre
print(mensaje)
```

### Longitud del texto
Esto cuenta caracteres incluyendo espacios.

```python
frase = "Hola a todos"
print(len(frase))
```

### Convertir mayúsculas y minúsculas

```python
texto = "Machine Learning"
print(texto.upper())  # MAYÚSCULAS
print(texto.lower())  # minúsculas
```

### Reemplazar palabras

```python
frase = "La IA es increíble"
nueva = frase.replace("increíble", "poderosa")
print(nueva)
```

### Quitar espacios al inicio y final

```python
sucio = "   Hola   "
limpio = sucio.strip()
print(limpio)
```

### Buscar dentro del texto

```python
texto = "Python es genial"
print("Python" in texto)  # True o False
print(texto.find("genial"))  # posición donde inicia
```

### Slicing (extraer partes)

```python
palabra = "Inteligencia"
print(palabra[0:4])   # Inte
print(palabra[-3:])   # cia
```

### Separar texto (split)

```python
frase = "IA datos modelos redes"
palabras = frase.split()  # separa por espacios
print(palabras)
```

### Unir lista en un string

```python
lista = ['IA', 'datos', 'modelos']
texto_unido = " - ".join(lista)
print(texto_unido)
```

# Listas
¿Qué es una lista?

Una lista es una colección ordenada de elementos.
Puede guardar números, texto, o cualquier tipo de dato.

```python
numeros = [10, 20, 30, 40]
nombres = ["Ana", "Luis", "María"]
mezcla = [1, "hola", 3.5, True]

print(mezcla)
```

## Acceder a elementos de la lista
Los elementos tienen posición (índice), empezando en 0.

```python
frutas = ["manzana", "pera", "uva", "mango"]

print(frutas[0])  # manzana
print(frutas[2])  # uva
print(frutas[-1]) # último elemento → mango
```

## Modificar elementos

```python
frutas[1] = "plátano"
print(frutas)
```

## Agregar elementos

```python
frutas.append("naranja")   # al final
frutas.insert(1, "kiwi")   # en posición 1
print(frutas)
```

## Eliminar elementos

```python
print('Original', frutas)
frutas.remove("uva")   # elimina por nombre
print(frutas)
frutas.pop(2)          # elimina por índice
print(frutas)
```

## Largo de una lista

```python
numeros = [4, 5, 9, 12]
print(len(numeros))
```

## Sumar, máximo, mínimo y promedio

```python
valores = [5, 10, 15, 20]
print(sum(valores))
print(max(valores))
print(min(valores))

promedio = sum(valores) / len(valores)
print("Promedio:", promedio)
```

## Ordenar listas

```python
nums = [10, 3, 7, 2, 9]
nums.sort()
print(nums)

nums.sort(reverse=True)
print(nums)
```

## Cortar listas (slicing)

```python
letras = ["a", "b", "c", "d", "e"]
print(letras[0:3])   # primeras 3
print(letras[2:])    # desde la posición 2 hasta el final
```

## Buscar elementos

```python
colores = ["rojo", "azul", "verde"]
print("rojo" in colores)  # True
print(colores.index("verde"))  # posición
```

# Diccionario
¿Qué es un diccionario?

Un diccionario guarda datos en formato clave → valor.
Sirve para representar objetos del mundo real:

* Un estudiante (nombre, edad, calificación)
* Un producto (precio, categoría)
* Información científica (valor, unidad, fecha)

```python
estudiante = {
    "nombre": "Joan",
    "edad": 27,
    "promedio": 95
}

print(estudiante)
type(estudiante)
```

## Acceder a los valores

```python
print(estudiante["nombre"])
print(estudiante["edad"])
```

## Modificar valores

```python
estudiante["promedio"] = 92
print(estudiante)
```

## Agregar nuevas claves y valores

```python
estudiante["ciudad"] = "Ensenada"
print(estudiante)
```

## Eliminar elementos

```python
estudiante.pop("edad")
print(estudiante)
```

## Revisar si una clave existe

```python
print("nombre" in estudiante)   # True
print("telefono" in estudiante) # False
```

## Recorrer diccionarios

Solo claves

```python
for clave in estudiante.keys():
    print(clave)
```

Solo valores

```python
for valor in estudiante.values():
    print(valor)
```

Claves y valores

```python
for clave, valor in estudiante.items():
    print(clave, ":", valor)
```

## Diccionarios dentro de listas
Muy común en ciencia de datos:

```python
productos = [
    {"nombre": "Laptop", "precio": 15000},
    {"nombre": "Mouse", "precio": 250},
    {"nombre": "Monitor", "precio": 3500},
]

# precio total
total = 0
for p in productos:
    total += p["precio"]

print("Total:", total)
```
